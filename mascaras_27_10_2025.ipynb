```python
# Cargar imagen en escala de grises
img = cv2.imread("foto.jpg", cv2.IMREAD_GRAYSCALE)

# Kernel de detección de bordes (Sobel horizontal)
kernel_sobel = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])
img_filtrada_sobel = cv2.filter2D(img, -1, kernel_sobel)

# Kernel de enfoque
kernel_enfoque = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])
img_filtrada_enfoque = cv2.filter2D(img, -1, kernel_enfoque)

# Kernel de desenfoque
kernel_desenfoque = np.array([
    [1, 1, 1],
    [1, 1, 1],
    [1, 1, 1]
])
img_filtrada_desenfoque = cv2.filter2D(img, -1, kernel_desenfoque)

# Kernel de detección de bordes simple
kernel_borders = np.array([
    [0, 0, 0],
    [-1, 1, 0],
    [0, 0, 0]
])
img_filtrada_borders = cv2.filter2D(img, -1, kernel_borders)

# Kernel de sharpen
kernel_sharpen = np.array([
    [1, -2, 1],
    [-2, 5, -2],
    [1, -2, 1]
])
img_filtrada_sharpen = cv2.filter2D(img, -1, kernel_sharpen)

# Kernel de detección de bordes (Norte)
kernel_norte = np.array([
    [1, 1, 1],
    [1, -2, 1],
    [-1, -1, -1]
])
img_filtrada_norte = cv2.filter2D(img, -1, kernel_norte)

# Kernel Gaussiano (simplificado)
kernel_gauss = np.array([
    [1, 2, 3,1,1],
    [2, 7, 11,7,2],
    [3, 11, 17,11,3],
    [2, 7, 11,7,1],
    [1, 2, 3,2,1]
])
img_filtrada_gauss = cv2.filter2D(img, -1, kernel_gauss)

# Kernel repujado
kernel_repujado = np.array([
    [-2, -1, 0],
    [-1, 1, 1],
    [0, 1, 2]
])
img_filtrada_repujado = cv2.filter2D(img, -1, kernel_repujado)

# Kernel este
kernel_este = np.array([
    [-1, 1, 1],
    [-1, -2, 1],
    [-1, 1, 1]
])
img_filtrada_este = cv2.filter2D(img, -1, kernel_este)

# Kernel de desenfoque
kernel_bordes2 = np.array([
    [0, 1, 0],
    [1, -4, 1],
    [0, 1, 0]
])
img_filtrada_bordes2 = cv2.filter2D(img, -1, kernel_bordes2)



# Mostrar resultados
plt.figure(figsize=(15, 10))

plt.subplot(4, 3, 1)
plt.imshow(img, cmap='gray')
plt.title("Imagen original")
plt.colorbar()

plt.subplot(4, 3, 2)
plt.imshow(img_filtrada_sobel, cmap='gray')
plt.title("Imagen filtrada (Sobel horizontal)")
plt.colorbar()

plt.subplot(4, 3, 3)
plt.imshow(img_filtrada_enfoque, cmap='gray')
plt.title("Imagen filtrada (Enfoque)")
plt.colorbar()

plt.subplot(4, 3, 4)
plt.imshow(img_filtrada_borders, cmap='gray')
plt.title("Imagen filtrada (Borders)")
plt.colorbar()

plt.subplot(4, 3, 5)
plt.imshow(img_filtrada_sharpen, cmap='gray')
plt.title("Imagen filtrada (Sharpen)")
plt.colorbar()

plt.subplot(4, 3, 6)
plt.imshow(img_filtrada_norte, cmap='gray')
plt.title("Imagen filtrada (Norte)")
plt.colorbar()

plt.subplot(4, 3, 7)
plt.imshow(img_filtrada_desenfoque, cmap='gray')
plt.title("Imagen filtrada (desenfoque)")
plt.colorbar()

plt.subplot(4, 3, 8)
plt.imshow(img_filtrada_gauss, cmap='gray')
plt.title("Imagen filtrada (Gauss)")
plt.colorbar()

plt.subplot(4, 3, 9)
plt.imshow(img_filtrada_repujado, cmap='gray')
plt.title("Imagen filtrada (repujado)")
plt.colorbar()

plt.subplot(4, 3, 10)
plt.imshow(img_filtrada_este, cmap='gray')
plt.title("Imagen filtrada (este)")
plt.colorbar()

plt.subplot(4, 3, 11)
plt.imshow(img_bn,  cmap='gray')
plt.title("Imagen bn")
plt.colorbar()

plt.subplot(4, 3, 12)
plt.imshow(img_filtrada_bordes2, cmap='gray')
plt.title("Imagen b2")
plt.colorbar()

plt.tight_layout()
plt.show()
```

```python
# Cargar imagen en color (RGB)
img_bgr = cv2.imread("dino.jpg")
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Separar canales
R, G, B = cv2.split(img_rgb)

# Kernel Sobel horizontal
kernel = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])

# Aplicar convolución a cada canal
R_filtrado = cv2.filter2D(R, -1, kernel)
G_filtrado = cv2.filter2D(G, -1, kernel)
B_filtrado = cv2.filter2D(B, -1, kernel)

# Recombinar imagen filtrada
img_filtrada = cv2.merge([R_filtrado, G_filtrado, B_filtrado])

# Mostrar resultado
plt.imshow(img_filtrada)
plt.title("Detección de bordes en imagen RGB")
plt.axis('off')
plt.show()
```

```python
# Cargar imagen en color (RGB)
img_bgr = cv2.imread("dragon.jpg")
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)

# Separar canales
R, G, B = cv2.split(img_rgb)

# Define kernels
kernel_sobel = np.array([
    [-1, 0, 1],
    [-2, 0, 2],
    [-1, 0, 1]
])

kernel_enfoque = np.array([
    [0, -1, 0],
    [-1, 5, -1],
    [0, -1, 0]
])

kernel_borders = np.array([
    [0, 0, 0],
    [-1, 1, 0],
    [0, 0, 0]
])

kernel_sharpen = np.array([
    [1, -2, 1],
    [-2, 5, -2],
    [1, -2, 1]
])

kernel_norte = np.array([
    [1, 1, 1],
    [1, -2, 1],
    [-1, -1, -1]
])

kernel_gauss = np.array([
    [1, 2, 3,1,1],
    [2, 7, 11,7,2],
    [3, 11, 17,11,3],
    [2, 7, 11,7,1],
    [1, 2, 3,2,1]
])

kernel_repujado = np.array([
    [-2, -1, 0],
    [-1, 1, 1],
    [0, 1, 2]
])

kernel_este = np.array([
    [-1, 1, 1],
    [-1, -2, 1],
    [-1, 1, 1]
])

kernel_desenfoque = np.array([
    [1, 1, 1],
    [1, 1, 1],
    [1, 1, 1]
])

kernel_bordes2 = np.array([
    [0, 1, 0],
    [1, -4, 1],
    [0, 1, 0]
])

def apply_kernel_to_rgb(img_rgb, kernel):
    R, G, B = cv2.split(img_rgb)
    R_filtered = cv2.filter2D(R, -1, kernel)
    G_filtered = cv2.filter2D(G, -1, kernel)
    B_filtered = cv2.filter2D(B, -1, kernel)
    return cv2.merge([R_filtered, G_filtered, B_filtered])

img_filtrada_sobel = apply_kernel_to_rgb(img_rgb, kernel_sobel)
img_filtrada_enfoque = apply_kernel_to_rgb(img_rgb, kernel_enfoque)
img_filtrada_borders = apply_kernel_to_rgb(img_rgb, kernel_borders)
img_filtrada_sharpen = apply_kernel_to_rgb(img_rgb, kernel_sharpen)
img_filtrada_norte = apply_kernel_to_rgb(img_rgb, kernel_norte)
img_filtrada_gauss = apply_kernel_to_rgb(img_rgb, kernel_gauss)
img_filtrada_repujado = apply_kernel_to_rgb(img_rgb, kernel_repujado)
img_filtrada_este = apply_kernel_to_rgb(img_rgb, kernel_este)
img_filtrada_desenfoque = apply_kernel_to_rgb(img_rgb, kernel_desenfoque)
img_filtrada_bordes2 = apply_kernel_to_rgb(img_rgb, kernel_bordes2)


plt.figure(figsize=(15, 10))

plt.subplot(4, 3, 1)
plt.imshow(img_rgb)
plt.title("Imagen original")
plt.axis('off')

plt.subplot(4, 3, 2)
plt.imshow(img_filtrada_sobel)
plt.title("Imagen filtrada (Sobel horizontal)")
plt.axis('off')


plt.subplot(4, 3, 3)
plt.imshow(img_filtrada_enfoque)
plt.title("Imagen filtrada (Enfoque)")
plt.axis('off')

plt.subplot(4, 3, 4)
plt.imshow(img_filtrada_borders)
plt.title("Imagen filtrada (Borders)")
plt.axis('off')

plt.subplot(4, 3, 5)
plt.imshow(img_filtrada_sharpen)
plt.title("Imagen filtrada (Sharpen)")
plt.axis('off')

plt.subplot(4, 3, 6)
plt.imshow(img_filtrada_norte)
plt.title("Imagen filtrada (Norte)")
plt.axis('off')

plt.subplot(4, 3, 7)
plt.imshow(img_filtrada_gauss)
plt.title("Imagen filtrada (Gauss)")
plt.axis('off')

plt.subplot(4, 3, 8)
plt.imshow(img_filtrada_repujado)
plt.title("Imagen filtrada (repujado)")
plt.axis('off')

plt.subplot(4, 3, 9)
plt.imshow(img_filtrada_este)
plt.title("Imagen filtrada (este)")
plt.axis('off')

plt.subplot(4, 3, 10)
plt.imshow(img_filtrada_desenfoque)
plt.title("Imagen filtrada (desenfoque)")
plt.axis('off')

plt.subplot(4, 3, 11)
plt.imshow(img_filtrada_bordes2)
plt.title("Imagen filtrada (kernel_bordes2)")
plt.axis('off')


plt.tight_layout()
plt.show()
```
